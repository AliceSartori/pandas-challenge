```python
# Dependencies
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Path to the CSV file 
Heroes_of_Pymoli = 'Resources/purchase_data.csv'

# Read Purchasing File and and store into Pandas DataFrame
df_Heroes_of_Pymoli = pd.read_csv(Heroes_of_Pymoli)
df_Heroes_of_Pymoli.head()
```

### Player Count

```python
#total number of players
Number_of_players = len(df_Heroes_of_Pymoli['SN'].unique())
Number_of_players 
```

### Purchasing Analysis (Total)

```python
#number of unique item
Number_of_unique_items = len(df_Heroes_of_Pymoli['Item ID'].unique())
Number_of_unique_items
```

```python
#Average of purchase price
Average_price = round(df_Heroes_of_Pymoli['Price'].mean(),2) 
amount = Average_price
currency = "${:,.2f}".format(amount)
print(currency)
```

```python
#Total number of purchases
Purchases_tot= df_Heroes_of_Pymoli['Purchase ID'].count()
Purchases_tot
```

```python
Revenue_tot = df_Heroes_of_Pymoli['Price'].sum()
amount_rev = Revenue_tot
currency_rev = "${:,.2f}".format(amount_rev)
print(currency_rev)
```

```python
#Summary DataFrame
Summary_Purchases = pd.DataFrame({"Total Number of Players": [Number_of_players],
                                 "Number of Unique Items": [Number_of_unique_items],
                                 "Average Price": [currency],
                                 "Number of purchases" : [Purchases_tot],
                                 "Total Revenue" : [currency_rev]})
Summary_Purchases
```

## Purchasing Analysis (Gender) 

```python
df_Heroes_of_Pymoli.head(10)
```

```python
#gender count
df_gender_count = df_Heroes_of_Pymoli.groupby('Gender')['Price', 'SN'].count()
df_gender_count_dropped= df_gender_count.drop(columns=['Price'])
df_gender_count_dropped

#Alternative solution
#Purchase_male = (df_Heroes_of_Pymoli[df_Heroes_of_Pymoli['Gender'] == "Male"])['Purchase ID'].count()
#Purchase_male
```

```python
#gender price mean
df_average_price = round(df_Heroes_of_Pymoli.groupby('Gender')['Price', 'SN'].mean(),2)
df_average_price["Price"] = df_average_price["Price"].map("${:.2f}".format)
df_average_price

#Average_Price_M = round((df_Heroes_of_Pymoli[df_Heroes_of_Pymoli['Gender'] == "Male"])['Price'].mean(),2)
```

```python
#total purchase value 
Total_purchase = df_Heroes_of_Pymoli.groupby('Gender')['Price', 'SN'].sum()
Total_purchase["Price"] = Total_purchase["Price"].map("${:.2f}".format)
Total_purchase


#Purchase_Other= (df_Heroes_of_Pymoli[df_Heroes_of_Pymoli['Gender'] == "Other / Non-Disclosed"])['Purchase ID'].count()
```

```python
#  Average Purchase Total per Person by Gender
Female = (df_Heroes_of_Pymoli[df_Heroes_of_Pymoli['Gender'] == "Female"])['Price'].sum()/number_of_female
Male=((df_Heroes_of_Pymoli[df_Heroes_of_Pymoli['Gender'] == "Male"])['Price'].sum()/number_of_male)
Undisclosed = (df_Heroes_of_Pymoli[df_Heroes_of_Pymoli['Gender'] == "Other / Non-Disclosed"])['Price'].sum()/number_of_undisclosed
```

```python
merge_df = pd.merge(df_gender_count_dropped, df_average_price, on="Gender")
merge_df
```

```python
merge_df2 = pd.merge(merge_df, Total_purchase, on="Gender")
merge_df2
```

```python
Average_total = {'Avg Total Purchase per Person': [Male, Female, Undisclosed]}

# creating a Dataframe object from dictionary with custom indexing 
df_gender = pd.DataFrame(Average_total, 
                         index = ['Male', 'Female', 'Other / Non-Disclosed']) 

df = df_gender.rename_axis('Gender')
df_dollar = df.applymap("${0:.2f}".format)
df_dollar
```

```python
#put them all together 
df_gender_total = pd.merge(merge_df2.rename(columns= {'SN': 'Purchase Count', 'Price_x': 'Average Purchase Price', 
                                                       'Price_y':'Total Purchase Value'}), 
                           df_dollar, 
                           on = "Gender")
df_gender_total
```

## Top Spenders

```python
df_top_spenders=df_Heroes_of_Pymoli[['SN','Price']]
df_top_spenders
```

```python
#pulled the data out using groupbyn and the function aggregate
df_top_spenders.groupby('SN').agg(['mean', 'count','sum'])
```

```python
df_price_groupby_sn = df_top_spenders.groupby('SN').agg(['count','mean','sum'])['Price'].sort_values(['sum'],ascending=False).head()

df_price_groupby_sn['mean'] = df_price_groupby_sn['mean'].round(2).map("${0:.2f}".format).head()

df_price_groupby_sn['sum'] = df_price_groupby_sn['sum'].map("${0:.2f}".format)
```

```python
df_price_groupby_sn.rename(columns={"count": "Purchase Count", 
                                    "mean":"Average Purchase Price",
                                    "sum": "Total Purchase Value" })
```

## Most Popular Items

```python
df_popular = df_Heroes_of_Pymoli[['Item ID', 'Item Name', 'Price',]]
df_popular
```

```python
df_top_popular = df_popular.groupby(['Item ID', 'Item Name']).agg(['count', 'mean', 'sum'])['Price'].sort_values(['count'],ascending=False).head()
df_top_popular

df_top_popular['mean'] = df_top_popular['mean'].round(2).map("${0:.2f}".format).head()

df_top_popular['sum'] = df_top_popular['sum'].map("${0:.2f}".format)
```

```python
df_top_popular.rename(columns={"count": "Purchase Count", 
                                "mean":"Item Price",
                                "sum": "Total Purchase Value"})
```
